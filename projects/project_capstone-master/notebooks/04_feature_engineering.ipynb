## Import Libraries and Read in Data:

```python
# Import Libraries:

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from sklearn.preprocessing import StandardScaler

import requests
```

```python
# Read data:

df = pd.read_csv('../data/outputs/02_merged_clean.csv', index_col='geo_id')
```

```python
df.head(2)
```

## Generate Features:

```python
# Generate 'Percent Other Race':

df['pct_other_minority'] = df['pct_american_ind'] + df['pct_hawaiian_pacisldr'] + df['pct_other_race']
df['pct_other_minority'].describe()
```

```python
# Generate 'Population Density':

df['pop_density'] = df['total_pop'] / df['total_housing_units']
df['pop_density'].describe()
```

```python
# Generate 'Percent Renter Occupied Units':

df['pct_renter_occupied'] = df['renter_occupied_units'] / df['total_housing_units']
df['pct_renter_occupied'].describe()
```

```python
df.info()
```

## Drop Features for Modeling:

```python
# Drop features unneeded for modeling, features highly correlated with other variavles, and/or used only for feature generation:

df = df.drop(columns=['pct_american_ind', 
   'pct_hawaiian_pacisldr', 
   'pct_other_race', 
   'total_pop', 
   'total_housing_units', 
   'occupied_housing_units', 
   'total_units_in_structure', 
   'pop_in_households', 
   'per_cap_income', 
   'avg_family_size', 
   'pct_male', 
   'total_households',
   'units_built_after_2014',
   'owner_occupied_units',
   'renter_occupied_units',
   'median_home_value',
   'pct_white'])
```

```python
df.info()
```

```python
# Heat Map Correlation Matrix:

plt.figure(figsize=(20,15))
sns.set(font_scale=1.2)
plt.title('Correlation Matrix', fontsize=18)
mask = np.triu(df.corr())
ax = sns.heatmap(df.corr(), annot=True, cmap='coolwarm', 
                 linewidths=.5, mask=mask);
```
