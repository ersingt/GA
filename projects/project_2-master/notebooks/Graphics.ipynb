## Graphic Generation:

### Import Libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

### Import Data:

```python
# Read in clean_engineered training data

df_train = pd.read_csv('../datasets/train_cleaned_engineered.csv')
```

```python
# Visualize int64 variables vs Sale Price:

int_dtypes = ['int64']

num_cols = []
for i in df_train.columns:
    if df_train[i].dtype in int_dtypes:
        num_cols.append(i)
    else:
        pass

# Using subplots instead of pairplot for readability

plt.subplots(figsize=(12, 120))
plt.subplots_adjust(right=1.5, top=1.5)
for i, feature in enumerate(list(df_train[num_cols]), 1):
    plt.subplot(len(list(num_cols)), 2, i)
    sns.scatterplot(x=feature, 
               y='SalePrice', 
               hue='SalePrice',
               palette='Blues',     
               data=df_train)
    plt.title(f'Sale Price vs {feature}', fontsize=13)
    plt.xlabel(f"{feature}", fontsize=13)
    plt.ylabel('SalePrice', fontsize=13);
```

```python
# Visualize Float Variables vs SalePrice

flo_dtypes = ['float64']
flo_cols = []
for i in df_train.columns:
    if df_train[i].dtype in flo_dtypes:
        flo_cols.append(i)
    else:
        pass

# Using subplots instead of pairplot for readability

plt.subplots(figsize=(12, 120))
plt.subplots_adjust(right=1.5, top=1.5)
for i, feature in enumerate(list(df_train[flo_cols]), 1):
    plt.subplot(len(list(flo_cols)), 3, i)
    sns.scatterplot(x=feature,
                    y='SalePrice',
               palette='Blues',
               hue='SalePrice',
               data=df_train)
    plt.title(f'Sale Price vs {feature}', fontsize=14)
    plt.xlabel(f"{feature}", fontsize=14)
    plt.ylabel('Sale Price', fontsize=14)
    
```

```python
# Box Plot
plt.figure(figsize=(9, 9))
sns.boxplot(x='Overall Qual', y='SalePrice', palette='Blues',data=df_train)
plt.xlabel('Overall Quality', fontsize=16)
plt.ylabel('Sale Price', fontsize=16)
plt.title('Overall Quality vs Sale Price', fontsize=16);
```

```python
# Visualize categorical variables:

object_dtypes = ['object'] # make seperate plots for integers and floats
cat_cols = []
for i in df_train.columns:
    if df_train[i].dtype in object_dtypes:
        cat_cols.append(i)
    else:
        pass
    
# Using subplots for readability    
# Bars: 

plt.subplots(figsize=(14, 120))
plt.subplots_adjust(right=1.5, top=1.5)
for i, feature in enumerate(list(df_train[cat_cols]), 1):
    plt.subplot(len(list(cat_cols)), 3, i)
    sns.countplot(x=feature,
               data=df_train)
    plt.title(f'Frequency of {feature}', fontsize=12)
    plt.xlabel(f"{feature}", fontsize=12)
    plt.ylabel('Frequency', fontsize=10);
```
