# Cleaning and Feature Engineering:

### Read in data:

```python
# Import libraries:

import pandas as pd
```

```python
# Read in Kaggle data:

fake = pd.read_csv('../data/Fake.csv')
real = pd.read_csv('../data/True.csv')
```

### Create Indicator Variable and Merge Datasets:

```python
# Create indicator columns to identify real and fake news:

fake['type'] = 1
real['type'] = 0
```

```python
# Merge data types:

df = fake.append(real)
df
```

```python
# Check type (y): Appears well-balanced. 

df['type'].value_counts()
```

### Drop Subject Column:

```python
# Drop 'subject' column:

df.drop(columns='subject')
```

### Drop Duplicates:

```python
# Deduplicate: None appear to exist.

df.drop_duplicates()
```

### Export Merged Data:

```python
# Export: 

df.to_csv('../data/merged_clean.csv')
```
