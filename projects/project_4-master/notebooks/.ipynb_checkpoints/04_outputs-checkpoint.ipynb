Create output graphics for Title vs Text Analysis:

```python
# Import Libraries:

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import plotly as iplot
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.corpus import stopwords
```

```python
# Read in data: 

fake = pd.read_csv('../data/fake.csv')
real = pd.read_csv('../data/true.csv')
```

```python
fake.head()
```

```python
real.head()
```

```python
# Define Lemmatizer: 

def split_into_lemmas(text):
    text = text.lower()
    lemmer = WordNetLemmatizer()
    return "".join([lemmer.lemmatize(word) for word in text])

# Code adapted from GA DSI Local Lesson: NLP
```

```python
# Define function to list most frequent words:

def get_top_n_words(corpus, n=None):
    vec = CountVectorizer(stop_words = 'english', 
                          preprocessor = split_into_lemmas, 
                          strip_accents = 'ascii'
                          ).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

# Code adapted from https://towardsdatascience.com/a-complete-exploratory-data-analysis-and-visualization-for-text-data-29fb1b96fb6a
```

```python
# Top Words:

fake_title_words = get_top_n_words(fake['title'], 15)
for word, freq in fake_title_words:
    print(word, freq)
```

```python
fake_text_words = get_top_n_words(fake['text'], 15)
for word, freq in fake_text_words:
    print(word, freq)
```

```python
real_title_words = get_top_n_words(real['title'], 15)
for word, freq in real_title_words:
    print(word, freq)
```

```python
real_text_words = get_top_n_words(real['text'], 15)
for word, freq in real_text_words:
    print(word, freq)
```

```python
# Make dataframe for fake title words:

fake_title_words = pd.DataFrame(fake_title_words, columns = ['ReviewText' , 'count'])

# Generate Graph for Fake title Words: 

sns.set(font_scale=2)
plt.figure(figsize=(8, 10))
ax4 = sns.barplot(data=fake_title_words, 
                  x='count', 
                  y = 'ReviewText',
                  color = 'Orange')
ax4.set(xlabel='Count', ylabel='Word', title='Top Words from Fake News Titles')
plt.show()
```

### Investigate BiGrams:

```python
# Generate list of top 20 Fake News Title BiGrams:

def get_top_n_bigram(corpus, n=None):
    vec = CountVectorizer(stop_words = 'english', 
                          preprocessor = split_into_lemmas, 
                          strip_accents = 'ascii',
                          ngram_range=(2, 2)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

common_words = get_top_n_bigram(fake['title'], 20)
for word, freq in common_words:
    print(word, freq)

# Code adapted from https://towardsdatascience.com/a-complete-exploratory-data-analysis-and-visualization-for-text-data-29fb1b96fb6a
```

```python
# Generate list of top 20 Real News Title BiGrams:

def get_top_n_bigram(corpus, n=None):
    vec = CountVectorizer(stop_words = 'english', 
                          preprocessor = split_into_lemmas, 
                          strip_accents = 'ascii',
                          ngram_range=(2, 2)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

common_words = get_top_n_bigram(real['title'], 20)
for word, freq in common_words:
    print(word, freq)

# Code adapted from https://towardsdatascience.com/a-complete-exploratory-data-analysis-and-visualization-for-text-data-29fb1b96fb6a
```
