# Modeling and Evaluation:

### Read in data:

```python
# Import Libraries:

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Read in merged data:

df = pd.read_csv('../data/merged_clean.csv')
```

## Use Title to predict Fake News:

### Define X and y: 
This presents an opportunity to explore how post title and text can predict fake news both seperately and combined. 

```python
X = df['title']
y = df['type']
```

```python
# Observe length of y:

y.shape
```

```python
# Train/Test Split on X,y:

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state=42)
```

### Create Lemmatizer:

```python
# Function to lemmatize text:

def split_into_lemmas(text):
    text = text.lower()
    lemmer = WordNetLemmatizer()
    return "".join([lemmer.lemmatize(word) for word in text])

# Code adapted from GA DSI Local Lesson: NLP
```

### Run Logistic Regression and CountVectorizer (Titles):

```python
# Define Pipeline:

pipe1 = Pipeline([
    ('cvec', CountVectorizer()),
    ('lr', LogisticRegression(solver = 'liblinear'))
    ])
```

```python
# Define parameters:

pipe1_params = {
    'cvec__max_features': [1000, 2000],                                      
    'cvec__ngram_range': [(1,1), (1,2), (2,2)],
    'cvec__strip_accents': ['ascii'],   
    'cvec__preprocessor': [split_into_lemmas, None],
    'cvec__stop_words':['english'],
    'lr__C': np.linspace(.001, 3, 5)
    }
```

```python
# Set up GridSearchCV:

gs1 = GridSearchCV(pipe1,
                  param_grid = pipe1_params,
                  n_jobs = -1,
                  cv = 5)
```

```python
# Fit GridSearch to training data:

print('Fitting models...')
gs1.fit(X_train, y_train)
print('Done.')
```

```python
# Score:

print(f'Training score: {gs1.score(X_train, y_train)}')
print(f'Test score    : {gs1.score(X_test, y_test)}')
print(f'Best params   : {gs1.best_params_}')
```

### Run Logistic Regression and CountVectorizer (Text):

#### Re-define X:

```python
X = df['text']
y = df['type']
```

```python
# Train/Test Split on X,y:

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state=42)
```

```python
# Define Pipeline:

pipe2 = Pipeline([
    ('cvec', CountVectorizer()),
    ('lr', LogisticRegression(solver = 'liblinear'))
    ])
```

```python
# Define Parameters:

pipe2_params = {
    'cvec__max_features': [2000],                                      
    'cvec__ngram_range': [(1,2)],
    'cvec__strip_accents': ['ascii'],   
    'cvec__preprocessor': [split_into_lemmas],
    'cvec__stop_words':['english'],
    'lr__C': np.linspace(.001, 3, 5)
    }
```

```python
# Set up GridSearchCV:

gs2 = GridSearchCV(pipe2,
                  param_grid = pipe2_params,
                  n_jobs = -1,
                  cv = 5)
```

```python
# Fit GridSearch to training data:

print('Fitting models...')
gs2.fit(X_train, y_train)
print('Done.')
print(f'Training score: {gs2.score(X_train, y_train)}')
print(f'Test score    : {gs2.score(X_test, y_test)}')
print(f'Best params   : {gs2.best_params_}')
```

```python
# Generate Coeffs mapped to words:

coefs = gs1.best_estimator_.named_steps.lr.coef_  

log_coefs = pd.DataFrame(gs1.best_estimator_.named_steps.lr.coef_ , 
                        columns=gs1.best_estimator_.named_steps.cvec.get_feature_names())

log_coefs

# Code adapted from GA DSI DC-CC11 Project 3 and the brilliance of Derek McAdam
```

```python
# Get top 10 estimators (coeffs) for Fake News:

top10_coef = log_coefs.T.sort_values(0, ascending=False).head(20)

# Get top 10 coeffs for Real News:

bottom10_coef = log_coefs.T.sort_values(0, ascending=False).tail(20)


# Append Data:

top20 = bottom10_coef.append(top10_coef)
top20 = pd.DataFrame(top20)
top20
```

```python
# Generate plot of strongest predictors of fake news by coefficient values:

sns.set(font_scale=2)
plt.figure(figsize=(8, 15))
ax = sns.barplot(data=top20, 
                  x=0, 
                  y = top20.index,
                  palette = 'twilight'
                 )
ax.set(xlabel='Coeff', ylabel='Word', title='Strongest Predictors of Real and Fake News by Title')
plt.show()
```
